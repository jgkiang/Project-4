```python
import pandas as pd
import numpy as np
```

```python
df_marvel = pd.read_csv('data/marvel-wikia-data.csv')
df_dc = pd.read_csv('data/dc-wikia-data.csv')
```

```python
df_marvel.head()
```

```python
print(df_marvel.shape)
print(df_dc.shape)
```

```python
df_marvel.drop(columns = ['page_id', 'urlslug', 'FIRST APPEARANCE'])
```

```python
df_dc_dropped = df_dc.drop(columns = ['page_id', 'urlslug', 'FIRST APPEARANCE'])
df_dc_dropped.head()
```

```python
df_dc_dropped.info()
```

```python
#df_dc_dropped =df_dc_dropped['APPEARANCES'].notna()
```

```python
df_dc_dropped = df_dc_dropped.dropna(subset=['APPEARANCES'])
```

```python
df_dc_dropped = df_dc_dropped[df_dc_dropped['APPEARANCES']>0]
```

```python
df_dc_dropped.info()
```

```python
df_dc_dropped = pd.to_numeric(df_dc_dropped(['APPEARANCES']).astype(int))
```

```python
df_dc_dropped['APPEARANCES'] = df_dc_dropped['APPEARANCES'].astype(np.int64)
```

```python
df_dc_dropped['YEAR'] = df_dc_dropped['YEAR'].astype(np.int64)
```

```python
df_dc_dropped.info()
```

```python
df_dc_dropped["GSM"] = df_dc_dropped["GSM"].fillna('Heterosexual')
```

```python
df_dc_dropped["EYE"].mode()
```

```python
mode_eye = df_dc_dropped["EYE"].mode().values[0]
```

```python
df_dc_dropped["EYE"] = df_dc_dropped["EYE"].fillna(mode_eye)
df_dc_dropped.head()
```

```python
df_dc_dropped.info()
```

```python
df_dc_dropped = df_dc_dropped.dropna(subset=['YEAR'])
```

```python
df_dc_dropped['YEAR'] = df_dc_dropped['YEAR'].astype(np.int64)
```

```python
df_dc_dropped.info()
```

```python
df_dc_dropped["HAIR"].mode()
```

```python
df_dc_dropped.groupby(["GSM","EYE"]).count()
```

```python
mode_hair = df_dc_dropped["HAIR"].mode().values[0]
```

```python
df_dc_dropped["HAIR"] = df_dc_dropped["HAIR"].fillna(mode_hair)
df_dc_dropped.head()
```

```python
df_dc_dropped.info()
```

```python
df_dc_dropped = df_dc_dropped.dropna(subset=['ALIGN'])
```

```python
df_dc_dropped.info()
```

```python
df_dc_dropped["ID"].mode()
```

```python
mode_id = df_dc_dropped["ID"].mode().values[0]
```

```python
df_dc_dropped["ID"] = df_dc_dropped["ID"].fillna(mode_id)
df_dc_dropped.head()
```

```python
df_dc_dropped.info()
```

```python
df_dc_dropped = df_dc_dropped.drop(columns=['ALIVE'])
```

```python
df_dc_dropped["SEX"] = df_dc_dropped["SEX"].fillna("Unknown")
```
